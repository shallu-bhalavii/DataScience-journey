```python

# Random student dataset 🌶️🍄
np.random.seed(42)  # For reproducibility

data = {
    'Name': ['Shallu', 'Aman', 'Riya', 'Kabir', 'Meera', 'Arjun', 'Neha','Alice', 'Bob', 'Charlie','shallu','rohan'],
    'Age': np.random.randint(18, 26, size=12),
    'Math': np.random.randint(30, 101, size=12),
    'Science': np.random.randint(30, 101, size=12),
    'English': np.random.randint(30, 101, size=12)
}

df = pd.DataFrame(data)
print(df)
```

```python
# ------adding new columns to the DataFrame------

# Total marks
df['Total'] = df[['Math', 'Science', 'English']].sum(axis=1)

# Average marks
df['Average'] = df[['Math', 'Science', 'English']].mean(axis=1)

# Result
df['Result'] = np.where(df['Average'] >= 50, 'Pass', 'Fail')

df
```

```python
# Students who passed
passed_students = df[df['Result'] == 'Pass']

# Top 3 students by average
top_students = df.sort_values(by='Average', ascending=False).head(3)
```

```python
# Summary
print(df.describe())

# Average marks in each subject
subject_avg = df[['Math', 'Science', 'English']].mean()

# Who scored the highest in Science?
top_science_student = df.loc[df['Science'].idxmax(), 'Name']

df
top_science_student
```

```python

# Filter students who scored more than 80 in Math
filtered_df = df[df['Math'] > 80]

# Find the youngest among them
youngest_student = filtered_df.loc[filtered_df['Age'].idxmin()]

print(youngest_student)
```

```python
# Calculate average score for each student
df['Average'] = df[['Math', 'Science', 'English']].mean(axis=1)

# Function to assign grade
def assign_grade(avg):
    if avg >= 80:
        return 'A'
    elif avg >= 65:
        return 'B'
    elif avg >= 50:
        return 'C'
    else:
        return 'F'

# Apply grading function
df['Grade'] = df['Average'].apply(assign_grade)
df
```
